# LangChain Setup Instructions

This notebook demonstrates how to set up and use LangChain with different providers.

## Required Packages
Run the installation cells below to install the necessary packages:
- `langchain-openai`: For OpenAI models (GPT-3.5, GPT-4)
- `langchain-anthropic`: For Anthropic models (Claude)
- `langgraph`: For building stateful, multi-actor applications

## API Keys
To use the agents with real models, you'll need to set up API keys:
- For Anthropic: Set `ANTHROPIC_API_KEY` environment variable
- For OpenAI: Set `OPENAI_API_KEY` environment variable
